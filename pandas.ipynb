0の出現回数が最も多い行・列を取得する

```python
import pandas as pd
from icecream import ic

data = [
    [3, 1, 2, 3, 4, 5],
    [3, 1, 0, 0, 7, 9],
    [3, 0, 0, 0, 3, 3],
    [3, 1, 0, 3, 3, 3],
    [3, 1, 2, 3, 3, 3],
]

df = pd.DataFrame(data)

row_0 = (df == 0).sum(axis=1)
ic(row_0)
ic(row_0.max())
ic(row_0.idxmax())
```

```python
from icecream import ic

data = [1, 2, 3, 4, 5, 6, 7]
max_data_length = 3

x = 0
y = 0
for i in range(0, len(data), max_data_length):
    x = i
    y = i + 3
    ic(x)
    ic(y)
    ic(data[x:y])
```

```python
import pandas as pd
from icecream import ic

input_file = "./sample_in_data/pandas/pandas_in.csv"
df = pd.read_csv(input_file)
max_data_length = 16300

with pd.ExcelWriter(input_file.replace("csv", "xlsx").replace("in", "out")) as writer:
    for i in range(0, len(df.columns), max_data_length):
        x = i
        y = i + max_data_length
        if (y - x) > len(df.iloc[:, x:y].columns):
            y = i + len(df.iloc[:, x:y].columns)
        df.iloc[:, x:y].to_excel(
            writer, sheet_name=str(x) + "_" + str(y), index=False, header=False
        )
```
