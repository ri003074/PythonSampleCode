```python
import numpy as np
```

```python
# 多次元配列を定義
# a = np.array([0, 1, 2])
a = np.array([[0, 1, 2], [3, 4, 5]])
a
```

```python
# 多次元配列の形を確認
a.shape
```

```python
# 次元数を確認
a.ndim
```

```python
# サイズ
a.size
```

```python
# 形を指定、要素0の配列
a = np.zeros((3, 3))
a
```

```python
# 形を指定、要素も指定
a = np.full((3, 3), 9)
a
```

```python
# 配列同士の計算

a = np.full((3, 3), 2)
b = np.full((3, 3), -1)
a * b
```

```python
a * -1
```

```python
# pythonの配列からnumpyの配列
lst = [0, 1, 2]
lst_np = np.array(lst)
lst_np
```

```python
# numpyの配列からpythonの配列
list(lst_np)
```

```python
# グラフ
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 500)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
# pythonの配列を演算する
data = [[0, 1], [2, 3]]
data_np = np.array(data)
data = [list(data) for data in list(data_np * -1)]
data
```
