```python
import matplotlib.pyplot as plt
import pandas as pd

data = {"tr": [57, 59], "tf": [61, 63], "data_rate": ["10G", "20G"]}
df = pd.DataFrame(data)
df = df.set_index("data_rate")

fig = plt.figure()
ax = fig.add_subplot(xmargin=0.5)
df.plot(ax=ax, style=["o", "o"])
ax.set_xticks([i for i in range(len(df.index))])
ax.set_yticks([40, 50, 60, 70, 80])
ax.set_ylabel("y_label")
plt.savefig("./sample_out_data/matplotlib/basic.png")
```
